### ДЗ № 1 

   ### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 
   ### a) Найти вероятность того, что все карты – крести. 
   ### б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз. 

Ответ:
   а) В колоде из 52 карт 4 масти по 13 карт. Вероятность вытащить первой карту крести составляет:
   $$ \frac{13}{52} $$
      и последующие три :
   $$ \frac{12}{51}, \frac{11}{50}, \frac{10}{49} $$
   
   
      Вероятность что все карты крести составляет :
   $$  \frac{13}{52} * \frac{12}{51} * \frac{11}{50} * \frac{10}{49} = 0.26\% $$
   
   
   б) Вероятность вытащить туз в первой карте и последующих трех составляеn, если в предудыщей итерации успех не достигнут:
   $$ \frac{4}{52} , \frac{4}{51}, \frac{4}{50}, \frac{4}{49} $$

### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно.
### Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?



Ответ: 

1. Необходимо найти кол-во сочетаний из трех цифр от 0 до 9
    
 ##   $$ C_n^k = \frac{9!}{3!(9-3)!} = 84 $$
 
2. Вероятность с первой попытки открыть дверь составляет:
 ## $$ \frac{1}{84} $$

### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали.
### Какова вероятность того, что все извлеченные детали окрашены?

Ответ:

Вероятность того, что все детали окрашены составляет:

## $$ \frac{9}{15} * \frac{8}{14} * \frac{7}{13} = 18.5 \% $$

### 4. В лотерее 100 билетов. Из них 2 выигрышных. 
### Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Ответ:
    1) Найдем сочетания по 2 пары из 100  билетов:
## $$ C_n^k = \frac{100!}{2!(100-2)!} = 4950  $$ 

Вероятность того, что два приобретенных билета окажутся выигрышными составляет:
## $$ \frac{1}{4950} $$
